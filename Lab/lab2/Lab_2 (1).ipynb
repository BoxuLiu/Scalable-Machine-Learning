```python
# [0, 5] to [-1, 1] || [MIN, MAX], TO [A, B]
# y = (ymax-ymin)*(x-xmin)/(xmax-xmin) + ymin：
k = (float)(1 - (-1)) / (5 - 0) # (ymax-ymin) / (xmax-xmin)

train = train.withColumn("target", (-1 + k * (col("lable") - 0)) ).drop("lable")
test = test.withColumn("target", (-1 + k * (col("lable") - 0))).drop("lable")
dev = dev.withColumn("target", (-1 + k * (col("lable") - 0)) ).drop("lable")
```

```python
from sentence_transformers import SentenceTransformer
from sentence_transformers import LoggingHandler
from sentence_transformers import models, losses, util
from sentence_transformers.evaluation import EmbeddingSimilarityEvaluator
from sentence_transformers.readers import InputExample
```

```python
def Create_examples(df):
    examples = []
    for row in df.collect():
        text_a = row['sentence1']
        text_b = row['sentence2']
        lable = row['target']
        examples.append((InputExample(texts = [text_a, text_b], label = lable)))
    return examples
```

```python
def add_to_samples(sent1, sent2, label):
    if sent1 not in train_data:
        train_data[sent1] = {'contradiction': set(), 'entailment': set(), 'neutral': set()}
    train_data[sent1][label].add(sent2)
```

```python
def add_to_samplesTest(sent1, sent2, label):
    if sent1 not in test_data:
        test_data[sent1] = {'contradiction': set(), 'entailment': set(), 'neutral': set()}
    test_data[sent1][label].add(sent2)
```

```python
train_data = {}
with gzip.open(nli_dataset_path, 'rt', encoding='utf8') as fIn:
    reader = csv.DictReader(fIn, delimiter='\t', quoting=csv.QUOTE_NONE)
    for row in reader:
        if row['split'] == 'train':
            sent1 = row['sentence1'].strip()
            sent2 = row['sentence2'].strip()

            add_to_samples(sent1, sent2, row['label'])
            add_to_samples(sent2, sent1, row['label'])  #Also add the opposite
```

```python
import random
train_samplesM2 = []
for sent1, others in train_data.items():
    if len(others['entailment']) > 0 and len(others['contradiction']) > 0:
        train_samplesM2.append(InputExample(texts=[sent1, random.choice(list(others['entailment'])), random.choice(list(others['contradiction']))]))
        train_samplesM2.append(InputExample(texts=[random.choice(list(others['entailment'])), sent1, random.choice(list(others['contradiction']))]))
```

```python
test_data = {}
with gzip.open(nli_dataset_path, 'rt', encoding='utf8') as fIn:
    reader = csv.DictReader(fIn, delimiter='\t', quoting=csv.QUOTE_NONE)
    for row in reader:
        if row['split'] == 'test':
            sent1 = row['sentence1'].strip()
            sent2 = row['sentence2'].strip()

            add_to_samplesTest(sent1, sent2, row['label'])
            add_to_samplesTest(sent2, sent1, row['label'])
```

```python
test_samplesM2 = []
for sent1, others in test_data.items():
    if len(others['entailment']) > 0 and len(others['contradiction']) > 0:
        test_samplesM2.append(InputExample(texts=[sent1, random.choice(list(others['entailment'])), random.choice(list(others['contradiction']))]))
        test_samplesM2.append(InputExample(texts=[random.choice(list(others['entailment'])), sent1, random.choice(list(others['contradiction']))]))
```
